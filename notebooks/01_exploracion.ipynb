```python
# Análisis Exploratorio de Datos (EDA)
Proyecto MIA - Detección de anomalías en infraestructura TI
#Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Cargar datos
df = pd.read_csv("../data/sample/ejemplo_dataset.csv", sep=";")
df.head()
#Informacion general
df.info()
df.describe()
#Valores unicos
df["metric"].unique()
df["instance"].unique()
#Conversion de fechas
df["timestamp_utc"] = pd.to_datetime(df["timestamp_utc"])
#Grafico
cpu = df[df["metric"] == "cpu_pct"]

plt.figure()
plt.plot(cpu["timestamp_utc"], cpu["value"])
plt.title("CPU Usage")
plt.xticks(rotation=45)
plt.show()

## Hallazgos

- Se observa variabilidad en el uso de CPU
- Existen picos que podrían representar anomalías
- Las métricas están organizadas por servidor y tipo
```
